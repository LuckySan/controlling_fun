# Grundlagen der Regenlungstechnik 

## 1. Einleitung
Beispiel: Heizungssystem
- Wohnzimmer -> es ist kalt -> wir wollen die Temperature passen regeln

**Ziel der Regelungstechnik**
- **Sollwert** ein System (Wohnzimmer) -> soll auf einen bestimmten Sollwert gehalten werden oder dorthin gebracht werden (Temperatur Wohnzimmer)
- **Stellgröße**: wird beeinflusst, um die Regelgröße (aktuelle Temperatur des Wohnzimmers) zu erreichen (Auf und Zu der Heizung)

### Nomenklatur
- Heizung: Aktor 
- Thermometer: Sensor - misst die Stellgröße
- manchmal: Störungen -> z.B. Fenster auf, dann wird es kälter

**Ein Regelkreis ist wie ein Kreislauf:**

1. Wir messen die aktuelle Temperatur (Regelgröße).
2. Wir vergleichen sie mit unserer Wunschtemperatur (Sollwert).
3. Der Regler (du selbst) entscheidet, was zu tun ist (Heizung auf- oder zudrehen).
4. Die Heizung setzt das um.
5. Die Temperatur ändert sich, und der Kreislauf beginnt von vorn.

# 2. Mathematische Nomenklatur
- **Sollwert (oder Referenzwert)**: Wird oft mit w(t) (für "Wert" oder "wunsch") oder r(t) (für "reference") bezeichnet. Das ist der Wert, den die Regelgröße erreichen soll.
- **Regelgröß (oder Istwert)**: Wird meist mit y(t) (für "output") bezeichnet. Das ist der tatsächlich gemessene Wert des Systems.
- **Stellgröße**: Wird oft mit u(t) (für "unabhängige Variable" oder "input") bezeichnet. Das ist die Größe, mit der wir das System beeinflussen.
- **Regelabweichung (oder Fehler)**: Das ist die Differenz zwischen Sollwert und Regelgröße. Sie wird häufig mit e(t) (für "error") bezeichnet: e(t)=w(t)−y(t). Dieser Fehler ist die Eingangsgröße für den Regler.
- **Störgröße**: Wird oft mit z(t) bezeichnet. Das sind unerwünschte Einflüsse auf das System.

### Regler
**Regler**: Der Regler selbst wird oft als Funktion oder Block G_t(s) (im Frequenzbereich) oder R dargestellt. Er erzeugt aus der Regelabweichung die Stellgröße.
**Regelstrecke (oder Prozess)**: Das ist das System, das wir regeln wollen. Es wird oft mit G_S(s) oder P dargestellt. Es empfängt die Stellgröße und gibt die Regelgröße aus.

### Stabile vs. instabile Ruhelage
- *stabile Ruhelage*: e.g ein Pendel, an der untersten Position -> bei kleinen Auslenkungen um die Ruhelage -> kehrt immer wieder in diese zurück 
- *instabile Ruhelage*: e.g ein Pendel, an der obersten Position -> bei kleinen Auslenkungen um die Ruhelage -> pendelt immer weiter weg von dieser
-> Roboter: Instabile Ruhelage
